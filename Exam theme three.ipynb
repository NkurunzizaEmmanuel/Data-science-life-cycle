### Theme Three 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
df = pd.read_csv('Theme_three.csv')
df
```

```python
df.info()
```

```python
df["Adult population"]= df["Adult population"].fillna(df["Adult population"]).median()
df["Adult population"]= df["Adult population"].astype(int)
```

```python
df.info()
```

```python
missing_counts = df.isnull().sum()
missing_percentage = (missing_counts / len(df)) * 100
print(missing_percentage)
```

```python
#dropping the columns with more than or equal to 82% data missing

df = df.loc[:, missing_percentage < 70]

```

```python
df.info()
```

```python
# Filling the missing values in object variables with mode
df.fillna(df.select_dtypes(include='object').mode().iloc[0], inplace=True)
df.info()
```

```python
df.head(10)
```

```python
# cleaning out the % sign from the data set
for col in df.columns:
    if df[col].dtype == 'object':  
        if df[col].str.contains('%', na=False).sum() > 0:  
            df[col] = df[col].str.replace('%', '', regex=True).astype(float)
```
